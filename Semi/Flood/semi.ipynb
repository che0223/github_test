```python
RainBool = pd.read_csv('./fldfct.csv', header=None, skiprows=1)
# 제목을 제외하고 불러오기

RainBool.columns = ['0', '장소', '시간', '발령']
RainBool = RainBool[['장소', '시간', '발령']]
RainBool

# 크롤링으로 좌표구해와야하나? -> 서울안에 들어간거만 추출하기
```

```python
df2 = pd.DataFrame(RainBool)

target_word = '해제'
# target_word2 = '한강'

# Define a custom filtering function
# any() : 하나라도 True인게 있으면 True
# all() : 모두 True여야 True 반환
def meets_criteria(row):
    contains_target_word = any(target_word in str(cell) for cell in row)
    #contains_target_word2 = any(target_word2 in str(cell) for cell in row)

    return not contains_target_word #and contains_target_word2

# Filter rows meeting the criteria
filtered_df2 = df2[df2.apply(meets_criteria, axis=1)]

filtered_df2
```

```python

filtered_dfs[2008]['일시']= filtered_dfs[2008]['일시'].astype('str')
dates = filtered_dfs[2008]['일시'].str.split(' ')




filtered_df2['시간']= filtered_df2['시간'].astype('str')
dates2 = filtered_df2['시간'].str.split(' ')


if dates.str.get(0) == dates2.str.get(0) :
    filtered_dfs['flood'] == True
else :
    filtered_dfs['flood'] == False
    
filtered_dfs['flood'] 



def compare_dates(row):
    date1 = row['일시']
    date2 = row['시간']
    
    if date1 == date2:
        return True
    else:
        return False

# Convert the date columns to string
filtered_dfs[2008]['일시'] = filtered_dfs[2008]['일시'].astype('str')
filtered_df2['시간'] = filtered_df2['시간'].astype('str')

# Create a temporary DataFrame with columns from both DataFrames
temp_df = pd.DataFrame()
temp_df['일시'] = dates.str.get(0) 
temp_df['시간'] = filtered_df2['시간']





# Apply the custom function to compare dates
temp_df['flood'] = temp_df.apply(compare_dates, axis=1)

# Assign the 'flood' column to the original DataFrame
filtered_dfs[2008]['flood'] = temp_df['flood']


'''
Ture값만 출력
mask1 = (filtered_dfs[2008].flood == True)
df_flood = filtered_dfs[2008].loc[mask1, :]
print(df_flood)
'''
```
